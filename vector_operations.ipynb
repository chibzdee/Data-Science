```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vectors(list_v, list_label, list_color):
    _, ax = plt.subplots(figsize=(10, 10))
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xticks(np.arange(-10, 10))
    ax.set_yticks(np.arange(-10, 10))



    plt.axis([-10, 10, -10, 10])
    for i, v in enumerate(list_v):
        sgn = 0.4 * np.array([[1] if i==0 else [i] for i in np.sign(v)])
        plt.quiver(v[0], v[1], color=list_color[i], angles='xy', scale_units='xy', scale=1)
        ax.text(v[0]-0.2+sgn[0], v[1]-0.2+sgn[1], list_label[i], fontsize=14, color=list_color[i]) 

    plt.grid()
    plt.gca().set_aspect('equal')
    plt.show()

v = np.array([[2], [5]])
plot_vectors([v], [f"$v$"], ['black'])

```

### SCALAR MULTIPLICATION

```python
plot_vectors([v, 2*v, -2*v], [f"$v$", f"$2v$", f"$-2v$"], ["black", "green", "blue"])
```

### SUM OF VECTORS

```python
v = np.array([[1], [3]])
u = np.array([[4], [-1]])

plot_vectors([v, u, np.add(v, u)], [f'$v$', f'$u$', f'$v + $u'], ['black', 'black', 'red'])
```

### Norm of a Vector

```python
print(f"Norm of a vector v is:", np.linalg.norm(v))
```

### Dot product of two vectors

```python
x = np.array([[1, -2, -5]])
y = np.array([[4, 3, -1]])
```

```python
def dot(x, y):
    s = 0
    for xi, yi in zip(x, y):
        s += xi * yi
    return s
```

```python
print('The dot product of x and y is:', dot(x[0], y[0]))
```

```python
#using np.dot()

np.dot(x[0], y[0])
```

### Geometric definition of dot product

```python
i = np.array([[1, 0, 1]])
j = np.array([[0, 1, 0]])


print(f"The dot product of i and j is: {dot(i[0], j[0])}")
```
